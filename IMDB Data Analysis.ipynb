```python
def rating (rating):
    if rating >=7.0:
        return('Excellent')
    elif rating>=6.0:
      return('Good')
    else:
      return('Average')
```

```python
top_rated_movies = df[['Title','Genre','Rating','Director']].sort_values(by='Rating', ascending=False).head(10)
top_rated_movies
```

```python
top_rated_movies.plot(kind='bar', x='Title', y='Rating', legend=False, title='Top-Rated Movies')
plt.ylabel('Rating')
plt.show()
```

```python
df['Rating'].plot(kind='hist', bins=20, title='Rating Distribution')
plt.xlabel('Rating')
plt.show()
```

```python
df['Genre'] = df['Genre'].str.split(',')
exploded_genres = df.explode('Genre')
genre_avg_ratings = exploded_genres.groupby('Genre')['Rating'].mean().sort_values(ascending=False)
genre_avg_ratings
```

```python
yearly_genre_trends = exploded_genres.groupby(['Year', 'Genre'])['Rating'].mean().unstack()
yearly_genre_trends[['Action', 'Drama', 'Comedy','Animation']].plot(title='Genre Rating Trends Over Time')
plt.ylabel('Average Rating')
plt.show()
```

```python
actors = df['Actors'].str.split(',')
actors = actors.apply(lambda x: [i.strip() for i in x])  
actors = actors.explode()
actors.value_counts().head(20)
```

```python
plt.figure(facecolor = "orange",figsize = (12, 6))
ax= sns.barplot(y = actors.value_counts().head(20).index, x = actors.value_counts().head(20).values)
ax.set(xlabel='Count',ylabel='Actors', title='Top 20 Actors by Number of Movies')
ax.set_facecolor("xkcd:mint green")
ax.bar_label(ax.containers[0])
plt.show()
```
